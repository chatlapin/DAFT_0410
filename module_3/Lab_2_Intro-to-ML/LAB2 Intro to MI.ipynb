```python
# isna() to select all rows with NaN under a single DataFrame column: df[df['column name'].isna()]
# isnull() to select all rows with NaN under a single DataFrame column: df[df['column name'].isnull()]

# Using isna() to select all rows with NaN under an entire DataFrame: df[df.isna().any(axis=1)]
# Using isnull() to select all rows with NaN under an entire DataFrame: df[df.isnull().any(axis=1)]
# ref: https://datatofish.com/rows-with-nan-pandas-dataframe/
```

# Challenge 3 - Processing the Events Column

```python
austin_fixed.Events.value_counts()
```

```python
#What is the largest number of events happened in a single day? Enter your answer in the next cell.
#Fog , Rain , Thunderstorm 
```

```python
event_list = ['Snow', 'Fog', 'Rain', 'Thunderstorm']

for event in event_list:
    austin_fixed[event] = 0

austin_fixed
```

```python
for event in event_list:
    austin_fixed[event] = np.where(austin_fixed.Events.str.contains(event), 1, 0)
```

```python
austin_fixed
```

```python
austin_fixed.drop(labels=['Events'], axis=1, inplace=True)
austin_fixed
```

# Challenge 4 - Processing The Date Column

```python
austin_fixed.Date = [dt.strptime(d, "%Y-%m-%d").toordinal() for d in austin_fixed.Date]
```

# Challenge 5 - Sampling and Holdout Sets

```python
features = ['TempHighF',
 'TempLowF',
 'DewPointHighF',
 'DewPointAvgF',
 'DewPointLowF',
 'HumidityHighPercent',
 'HumidityAvgPercent',
 'HumidityLowPercent',
 'SeaLevelPressureHighInches',
 'SeaLevelPressureAvgInches',
 'SeaLevelPressureLowInches',
 'VisibilityHighMiles',
 'VisibilityAvgMiles',
 'VisibilityLowMiles',
 'WindHighMPH',
 'WindAvgMPH',
 'WindGustMPH',
 'Snow',
 'Fog',
 'Rain',
 'Thunderstorm']

y = austin_fixed.TempAvgF
X = austin_fixed[features]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# ref: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.8)
```
