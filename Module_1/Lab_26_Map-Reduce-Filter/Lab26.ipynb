```python
prophet_cut = prophet[568:]
print(len(prophet_cut))
```

```python
print(prophet_cut[1:11])
```

```python
def reference(x):
    return x.split('{')[0]
```

```python
prophet_reference = list(map(reference, prophet_cut))
```

```python
def line_break(x):
    return x.split('\n')
```

```python
prophet_line = list(map(line_break, prophet_reference))
```

```python
prophet_flat = [item for sublist in prophet_line for item in sublist]
prophet_flat
```

```python
def word_filter(x):
    word_list = ['and', 'the']
    counting = sum([x.count(i) for i in word_list])
    return False if counting > 0 else True
```

```python
prophet_filter = list(filter(word_filter, prophet_flat))
prophet_filter
```
