```python
df.loc[df['Name'].str.contains("Mega"),'Name'] = 'Mega' + df['Name'][df['Name'].str.contains("Mega")].str.split
#df['Name'].apply(lambda x: x.replace('*Mega*', ''))
df.head(10)
```

```python
df['A/D Ratio'] = df['Attack']/df['Defense']
df['A/D Ratio']
```

```python
max_Pokemon=df['A/D Ratio'].max()
df[df['A/D Ratio']==max_Pokemon]
```

```python
min_Pokemon=df['A/D Ratio'].min()
df[df['A/D Ratio']==min_Pokemon]
```

```python
df.loc[df['Type 2'].isnull()==False, 'Combo Type']=df[['Type 1', 'Type 2']].astype(str).apply('-'.join, axis=1)
df.loc[df['Type 2'].isnull()==True, 'Combo Type']=df['Type 1']
```
